```python
import matplotlib.pyplot as plt
import cv2
%matplotlib inline

import dxfgrabber

dxf = dxfgrabber.readfile("test_extensor_hood_v1.dxf")
print(dxf.entities)
```

```python
circles = [e for e in dxf.entities if e.dxftype == 'CIRCLE']
lines = [e for e in dxf.entities if e.dxftype == 'LINE']
points = [e for e in dxf.entities if e.dxftype == 'POINT']
```

```python
cirs[0].center
```

```python
for circle in circles:
    print(circle.center)
    x = circle.center[0]
    y = circle.center[1]
    plt.plot(x, y, marker='.', markersize=20)
```

```python
for line in lines:
    #print(line.start, line.end)
    ls_x, ls_y = line.start[0], line.start[2]
    le_x, le_y = line.end[0], line.end[2]
    plt.plot([ls_x, le_x], [ls_y, le_y], 'k-')
```

```python
for point in points:
    print(point.point)
    p_x, py = point.point[0], point.point[2]
    plt.plot(p_x, py, marker='.', markersize=20)
```

```python
plt.figure(figsize=(10, 3))

lst_ls, lst_le = [], []

for line in lines:
    #print(line.start, line.end)
    ls_x, ls_y = line.start[0], line.start[2]
    le_x, le_y = line.end[0], line.end[2]
    plt.plot([ls_x, le_x], [ls_y, le_y], 'k-')
    
    plt.plot(line.start[0], line.start[2], marker='v', markersize=5, color='black')
    plt.plot(line.end[0], line.end[2], marker='v', markersize=5, color='black')
    
    lst_ls.append([line.start[0], line.start[2]])
    lst_le.append([line.end[0],  line.end[2]])
    
for circle in circles:
    x = circle.center[0]
    y = circle.center[2]
    plt.plot(-x, y, marker='.', markersize=20, color='red')
```

```python
for start_xy in lst_ls:
    for end_xy in lst_le:
        if start_xy == end_xy:
            print(start_xy)
```
