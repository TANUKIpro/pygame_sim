```python
import matplotlib.pyplot as plt
%matplotlib inline
import dxfgrabber

dxf = dxfgrabber.readfile("test_extensor_hood_v2.dxf")
print(dxf.entities)
```

```python
circles = [e for e in dxf.entities if e.dxftype == 'CIRCLE']
lines = [e for e in dxf.entities if e.dxftype == 'LINE']
points = [e for e in dxf.entities if e.dxftype == 'POINT']
```

```python
circles[0].center
```

```python
for circle in circles:
    #print(circle.center)
    x = circle.center[0]
    y = circle.center[1]
    plt.plot(-x, y, marker='.', markersize=10, color="orange")
```

```python
for line in lines:
    #print(line.start, line.end)
    ls_x, ls_y = line.start[0], line.start[2]
    le_x, le_y = line.end[0], line.end[2]
    plt.plot([ls_x, le_x], [ls_y, le_y], 'k-')
```

```python
for point in points:
    #print(point.point)
    p_x, p_y = point.point[0], point.point[2]
    plt.plot(p_x, p_y, marker='.', markersize=20)
```

```python
plt.figure(figsize=(10, 3))

lst_ls, lst_le = [], []
for line in lines:
    #print(line.start, line.end)
    ls_x, ls_y = -1*line.start[0]*10+10, line.start[2]*10+50
    le_x, le_y = -1*line.end[0]*10 +10, line.end[2]*10+50
    plt.plot([ls_x, le_x], [ls_y, le_y], 'k-')
    
    plt.plot(ls_x, ls_y, marker='.', markersize=5, color='black')
    plt.plot(le_x, le_y, marker='.', markersize=5, color='black')
    
    lst_ls.append([line.start[0], line.start[2]])
    lst_le.append([line.end[0],  line.end[2]])
    
for circle in circles:
    x = circle.center[0]*10+10
    y = circle.center[1]*10+50
    
    x = round(x, 3)
    y = round(y, 3)

    if x == -7 or x == 96:
        plt.plot(x, y, marker='.', markersize=15, color="green")
    else:
        plt.plot(x, y, marker='.', markersize=7, color="red")

#for point in points:
    #p_x, p_y = point.point[0], point.point[2]
    #plt.plot(p_x, p_y, marker='.', markersize=20, )
```
